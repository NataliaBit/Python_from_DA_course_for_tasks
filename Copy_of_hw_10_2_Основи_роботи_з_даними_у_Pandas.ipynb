Весь аналіз даних будується на тому, що ми перевіряємо певні гіпотези та робимо висновки і рекоемндації на основі своїх досліджень.

```python
yangon_total=df[df.City =='Yangon']['Total'].mean()
mandalay_total=df[df.City =='Mandalay']['Total'].mean()

yangon_total,mandalay_total

```

Можемо побачити, що в місті Mandalay більше і gross income, і привязаний до нього показник Total (загальна сума покупок), що також є важливим показником, тому в аналізі будемо спиратися і на нього. Водночас бачимо, що кількість інвойсів в Mandalay менша - 332 проти 340 в Yangon, тому вищий показник загальних продажів в Mandalay, можна зробити висновок, - досягається за рахунок більших сум чеків.
Тому маємо 3 гіпотези, чим більший Total в місті Mandalay досягається:
1) більшими сумами продажів загалом - тобто більшими середніми чеками, ніж в місті Yangon (наприклад, більшою кількістю товарів Quantity в чеках, комбінацією кількох категорій товарів в одній покупці)
2) більшими цінами на товари в місті Mandalay - вищий Unit price- і як результат, більшими сумами чеків; продаж більш дорогих категорій товарів - що автоматично також підвищує суму середнього чеку
3) більшою часткою крупних чеків (покупок з високою сумою), що зміщує середній Total вгору. Якщо говорити про розподіл чеків Total в даному випадку -то можна сказати про більший правий хвост розподілу (це можна буде побачити також пізніще в значенні персентилів).


```python
yangon_quantity = df[df.City=='Yangon']['Quantity'].mean()
mandalay_quantity = df[df.City=='Mandalay']['Quantity'].mean()

yangon_quantity,mandalay_quantity
```

```python
yangon_unprice= df[df.City=='Yangon']['Unit price'].mean()
mandalay_unprice= df[df.City=='Mandalay']['Unit price'].mean()

yangon_unprice,mandalay_unprice
```

Порахуємо кількість товарів в чеках (середню) та середні ціни на товари в містах Mandalay та Yangon для перевірки наших перших двох гіпотез. Як бачимо, в обоз мітсах кількість товарі в чеках в середньому не має значущої різниці і майже ожнакова. Тому першу гіпотезу можемо відкинути.
Тоді як середня ціна на товари трохи вища - і саме вона може давати різницю в чеках в більший бік. Тому другу гвпотезу поки залищаємо.
