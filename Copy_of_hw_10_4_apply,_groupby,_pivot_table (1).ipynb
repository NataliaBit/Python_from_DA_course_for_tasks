Таким чином, як висновок по гіпотезі 1 можна зробити висновок: дохід по філіях А і В не мав особливих відмінностей залежно від статі та членства, але у філії С продажі з найвищим доходом приносять жінки з членством карток лояльності магазину.
```python
# Дослідимо, товари яких категорій  в яких філіях продавалися найбільше
products_max = pd.pivot_table(
    df,
    index='Branch',
    columns=['Product line'],
    values='Total',
    aggfunc='mean'
)

products_max
```

Лідирує знову майже в усіх категоріях по продажах філія С; дослідимо вартість товарів в філіях для порівняння ( чи не дорожче товари в С?),середній чек та кількість товарів в чеку

```python
# І також обчислимо середню кількість товарів в чеку по філіях
avg_quantity_by_branch = df.groupby('Branch')['Quantity'].mean()
avg_quantity_by_branch
```

3. Обчислимо тепер, в який час були зроблені найбільші продажі, чи інсує тут залежність і чи є звязок з попередніми двома гіпотезами. Наприклад, чи є певні дні, коли жінки купували найбільше і яких саме товарів.

```python
# Подивимося, в який день тижня в філіях маємо найбільше продажів
df['Date'] = pd.to_datetime(df['Date'])
df['day_of_week'] = df['Date'].dt.day_name()
```

```python
day_of_week_revenue = pd.pivot_table(
    df,
    index='Branch',
    columns='day_of_week',
    values='Total',
    aggfunc='sum'
)

day_of_week_revenue
```

```python
day_of_week_revenue.idxmax(axis=1)
```

Найбільше продажів філії мають очікувано у вихдні дні- в суботу та неділю.

```python
# Перевіримо звязок наявності карток лояльності покупців з покупками у певні дні тижня
day_of_week_gender=pd.pivot_table(
    df,
    index='day_of_week',
    columns=['Customer type'],
    values='Total',
    aggfunc='sum'
)
day_of_week_gender
```

```python
day_of_week_gender.idxmax(axis=1)
```

ЯК бачимо,покупок здійснюють більше мембери, причому в тому числі у вихідні ця тенденція зберігається

```python
# Перевіримо, чи в філіяїх існує звязок між покупками в певні дні тижня та статтю
day_of_week_gender=pd.pivot_table(
    df,
    index='day_of_week',
    columns=['Gender','Branch'],
    values='Total',
    aggfunc='sum'
)
day_of_week_gender
```

```python
day_of_week_gender.idxmax(axis=1)
```

Цікава тенденція, що найбільше покупок здійснюють саме жінки, майже в усі дні тижня, найчастіше знову ж таки у філії С.

Висновок: як бачимо, філія C демонструє найвищі фінансові показники завдяки тому, що основний дохід її формують жінки, зокрема учасниці програми лояльності (Members). вищої активності жінок, ефективності membership-моделі та підвищеного попиту у вихідні. Ці показники необхідно ретельно вивчити задля можливого перенесення успіху філії С на інші гілки
